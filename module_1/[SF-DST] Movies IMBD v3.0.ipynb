# 30. Какой месяц чаще всего по годам самый прибыльный?
Варианты ответа:
1. Январь
2. Июнь
3. Декабрь
4. Сентябрь
5. Май

```python
s = set()
for date in data['release_year']:
    s.add(date)

pivot = data.pivot_table(values=['profit'],
index=['release_year'],
columns=['month'],
aggfunc='sum')
display(pivot)

best_month_array = []
bm = set()
for year in s:
    year_best_month = pivot.loc[year].idxmax()[1]
    best_month_array.append(year_best_month)
    bm.add(year_best_month)
for month in bm:
    print(month ,best_month_array.count(month))
```

# 31. Названия фильмов какой студии в среднем самые длинные по количеству символов?
Варианты ответа:
1. Universal Pictures (Universal)
2. Warner Bros
3. Jim Henson Company, The
4. Paramount Pictures
5. Four By Two Productions

```python
mask = data['production_companies'].str.contains('Four By Two Productions')
data[mask]
```

```python
data['title_len'] = data['original_title'].str.len()

companies = set()
companies_series = data['production_companies'].str.split('|')
for row in companies_series:
    for studio in row:
        companies.add(studio)
        
mean_array = []
for company in companies:
    mask = data['production_companies'].str.contains(company, regex = False)
    mean_title = data[mask]['title_len'].mean()
    mean_array.append(mean_title)
result = pd.Series(data=mean_array, index=companies)
print(result.sort_values().tail(50))
print(len(companies), len(mean_array))
```

# 36. У какого из режиссеров выше вероятность выпустить фильм в прибыли? (5 баллов)101
Варианты ответа:
1. Quentin Tarantino
2. Steven Soderbergh
3. Robert Rodriguez
4. Christopher Nolan
5. Clint Eastwood

```python
def hip(hop):
    if hop:
        return 1
    return 0
    
data['is_profit'] = (data['profit']>0).apply(hip)

directors = ['Quentin Tarantino', "Steven Soderbergh", 'Robert Rodriguez', 'Christopher Nolan', 'Clint Eastwood']
for diror in directors:
    mask = data['director'].str.contains(diror)
    result = data[mask]['is_profit'].mean()
    print(diror, result)
```
